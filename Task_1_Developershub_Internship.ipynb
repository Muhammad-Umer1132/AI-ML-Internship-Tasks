**Task 1: Exploring and Visualizing a Simple Dataset**

**Problem Statement:**  The goal of this task is to load,inspect and visualize the famous **Iris Dataset** to understand data trends, distributions and relationships between features. This forms the foundation of any data science workflow - understanding your data before building any model.

**DataSet:** The Iris Dataset contains 150 samples of iris flowers from 3 species;
**i. Iris Setosa,**
**ii. Iris Versicolor,**
**iii. Iris Virginica.**
Each Sample has 4 features; **i. Sepal Length, ii. Sepal Width,** **iii. Petal Length and Petal Width (all in cm).**


**Objective:** Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.

**Step 1: Import Required Libraries**

```python
# Core data manipulation library
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# To load the Iris dataset directly
from sklearn.datasets import load_iris

# Display plots inline in the notebook
%matplotlib inline

# Set a clean visual style for all seaborn plots
sns.set_theme(style='whitegrid', palette='Set2')
plt.rcParams['figure.dpi'] = 120

print("✅ All libraries imported successfully!")
```

**Step 2: Load The DataSet**

```python
# Load the Iris dataset from sklearn (same data as the CSV version)
iris_data = load_iris()

# Convert it into a pandas DataFrame for easy manipulation
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

# Add the target (species) column — 0=Setosa, 1=Versicolor, 2=Virginica
df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)

# Rename columns for cleaner display
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

print("✅ Dataset loaded successfully!")
print(f"Dataset shape: {df.shape}")
```

**Step 3: Initial Data Inspection**

```python
# 3.1 — Shape: number of rows and columns
print(" Dataset Shape (rows, columns):")
print(df.shape)
print()
```

```python
# 3.2 — Column Names
print(" Column Names:")
print(df.columns.tolist())
```

```python
# 3.3 — First 5 rows of the dataset
print("First 5 Rows:")
df.head()
```

```python
# 3.4 — Dataset Info: data types and null values
print(" Dataset Info:")
df.info()
```

```python
# 3.5 — Descriptive Statistics
print("Descriptive Statistics:")
df.describe()
```

```python
# 3.6 — Check for Missing Values
print("Missing Values per Column:")
print(df.isnull().sum())
print()
print("No missing values found! The dataset is clean.")
```

```python
# 3.7 — Class Distribution
print("Species Distribution (samples per class):")
print(df['species'].value_counts())
```

**Insight: The dataset is perfectly balanced.**

**Step 4: Data Visualization**

**4.1 - Scatter Plots: Feature Relationships**

```python
# Pairplot — shows scatter plots between every pair of features
# This helps us understand which features best separate the species

plt.figure(figsize=(2, 2))
pair_plot = sns.pairplot(
    df,
    hue='species',           # Color points by species
    diag_kind='kde',         # Diagonal: Kernel Density Estimate
    markers=['o', 's', 'D'], # Different markers per species
    plot_kws={'alpha': 0.7}
)
pair_plot.fig.suptitle('Pairplot of Iris Features by Species', y=1.02, fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('scatter_pairplot.png', bbox_inches='tight')
plt.show()
print("Pairplot saved as scatter_pairplot.png")
```

```python
# Focused Scatter Plot: Petal Length vs Petal Width (best separator)
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x='petal_length',
    y='petal_width',
    hue='species',
    style='species',
    s=100,
    alpha=0.8
)
plt.title('Petal Length vs Petal Width by Species', fontsize=13, fontweight='bold')
plt.xlabel('Petal Length (cm)', fontsize=11)
plt.ylabel('Petal Width (cm)', fontsize=11)
plt.legend(title='Species')
plt.tight_layout()
plt.savefig('scatter_petal.png', bbox_inches='tight')
plt.show()
print("Scatter plot saved as scatter_petal.png")
```

**4.2 - Histograms: Value Distributions**

```python
# Histograms for all 4 numeric features, split by species
# This helps us understand how each feature is distributed within each class

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
titles = ['Sepal Length (cm)', 'Sepal Width (cm)', 'Petal Length (cm)', 'Petal Width (cm)']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Feature Value Distributions by Species', fontsize=14, fontweight='bold')

for ax, feature, title in zip(axes.flatten(), features, titles):
    # Plot one histogram per species, overlaid on same axis
    for species in df['species'].unique():
        subset = df[df['species'] == species]
        ax.hist(subset[feature], bins=15, alpha=0.6, label=species, edgecolor='white')

    ax.set_title(title, fontsize=11, fontweight='bold')
    ax.set_xlabel('Value (cm)')
    ax.set_ylabel('Frequency')
    ax.legend(title='Species')

plt.tight_layout()
plt.savefig('histograms.png', bbox_inches='tight')
plt.show()
print("Histograms saved as histograms.png")
```

**4.3 - Box Plots: Outlier Detection**

```python
# Box plots reveal the median, spread (IQR), and outliers (dots outside whiskers)
# We plot each feature grouped by species

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Box Plots: Feature Distribution & Outlier Detection by Species', fontsize=14, fontweight='bold')

for ax, feature, title in zip(axes.flatten(), features, titles):
    sns.boxplot(
        data=df,
        x='species',
        y=feature,
        ax=ax,
        palette='Set2',
        width=0.5,
        flierprops=dict(marker='o', color='red', markersize=6, label='Outlier')
    )
    ax.set_title(title, fontsize=11, fontweight='bold')
    ax.set_xlabel('Species')
    ax.set_ylabel('Value (cm)')

plt.tight_layout()
plt.savefig('boxplots.png', bbox_inches='tight')
plt.show()
print("Box plots saved as boxplots.png")
```

**4.4 - Extra(Heatmap)**

```python
# Heatmap shows how strongly each pair of features is correlated
# Values close to 1 or -1 indicate strong correlation

plt.figure(figsize=(7, 5))
corr_matrix = df[features].corr()  # Compute Pearson correlation

sns.heatmap(
    corr_matrix,
    annot=True,           # Show correlation values
    fmt='.2f',            # 2 decimal places
    cmap='coolwarm',      # Color: red=positive, blue=negative correlation
    linewidths=0.5,
    square=True,
    cbar_kws={'shrink': 0.8}
)
plt.title('Feature Correlation Heatmap', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('heatmap.png', bbox_inches='tight')
plt.show()
print("✅ Heatmap saved as heatmap.png")
```
